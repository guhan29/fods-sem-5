# plotting multiple lines in one cell

```python
plt.figure(figsize=(14,6))
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.subplot(1,2,1) #subplot(1 row, 2 column, #plot number) set the position with subplot first
plt.plot(x,y1,color = 'yellow')#rgbcmyk
plt.xlabel('X')
plt.ylabel('Sin(x)')
plt.subplot(1,2,2)
plt.plot(x,y2,'b')
plt.xlabel('X')
plt.ylabel('Cos(x)')
plt.show()
```

```python
plt.figure(figsize=(14,6))
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.subplot(2,1,1) #subplot(1 row, 2 column, #plot number) set the position with subplot first
plt.plot(x,y1,color = 'yellow')#rgbcmyk
plt.xlabel('X')
plt.ylabel('Sin(x)')
plt.subplot(2,1,2)
plt.plot(x,y2,'b')
plt.xlabel('X')
plt.ylabel('Cos(x)')
plt.show()
```

# creating plots in grid

```python
x = np.arange(-50,50)
y1 = np.power(x,2)
y2 = np.power(x,3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)

fig1, ax1 = plt.subplots(nrows=3,ncols=3,figsize=(20,20))
ax1[0,0].plot(x,y1,"tab:blue")
ax1[0,0].set_title('Square Function')
ax1[0,0].set_xlabel(r'$x$',fontsize=18)
ax1[0,0].set_ylabel(r'$y$',fontsize=18)

ax1[0,1].plot(x,y2,"tab:red")
ax1[0,1].set_title('Cubic Function')
ax1[0,1].set_xlabel(r'$x$',fontsize=18)
ax1[0,1].set_ylabel(r'$y$',fontsize=18)

ax1[0,2].plot(x,y3,"tab:green")
ax1[0,2].set_title('Sine Function')
ax1[0,2].set_xlabel(r'$x$',fontsize=18)
ax1[0,2].set_ylabel(r'$y$',fontsize=18)

ax1[1,0].plot(x,y4,"tab:green")
ax1[1,0].set_title('Cosine Function')
ax1[1,0].set_xlabel(r'$x$',fontsize=18)
ax1[1,0].set_ylabel(r'$y$',fontsize=18)

ax1[1,1].plot(x,y5,"tab:green")
ax1[1,1].set_title('Tangent Function')
ax1[1,1].set_xlabel(r'$x$',fontsize=18)
ax1[1,1].set_ylabel(r'$y$',fontsize=18)

ax1[1,2].plot(x,y6,"tab:green")
ax1[1,2].set_title('Tanh Function')
ax1[1,2].set_xlabel(r'$x$',fontsize=18)
ax1[1,2].set_ylabel(r'$y$',fontsize=18)

ax1[2,0].plot(x,y7,"tab:green")
ax1[2,0].set_title('Sinh Function')
ax1[2,0].set_xlabel(r'$x$',fontsize=18)
ax1[2,0].set_ylabel(r'$y$',fontsize=18)

ax1[2,1].plot(x,y8,"tab:green")
ax1[2,1].set_title('Cosh Function')
ax1[2,1].set_xlabel(r'$x$',fontsize=18)
ax1[2,1].set_ylabel(r'$y$',fontsize=18)

ax1[2,2].plot(x,y9,"tab:green")
ax1[2,2].set_title('Exponential Function')
ax1[2,2].set_xlabel(r'$x$',fontsize=18)
ax1[2,2].set_ylabel(r'$y$',fontsize=18)
plt.show()
```

```python
col1 = np.linspace(0,10,1000)
col2 = np.sin(col1)
df = pd.DataFrame({'C1':col1,'C2':col2})
df.head(10)
```

```python
#%matplotlib inline
sns.lineplot(x=df.C1,y=df.C2,data=df)
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.set(rc={'axes.facecolor':'#FFFDE7','axes.grid':True,'xtick.labelsize':20,'ytick.labelsize':20})
sns.lineplot(x=df.C1,y=df.C2,data=df,linewidth=2)
sns.lineplot(x=df.C1,y=df.C2+1,data=df,linewidth=2)
plt.show()
```
