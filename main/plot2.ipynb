```python
import pandas as pd 
import numpy as np
import matplotlib as mpl 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.figure(figsize=(14,6))
x=np.linspace(0,10,1000)
y1=np.sin(x)
y2=np.cos(x)
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)
```

```python
x = np.arange(-50,50)
y1 = np.power(x,2)
y2 = np.power(x,3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)

y6 = np.tanh(x)
y7 = np.cosh(x)
y8 = np.sinh(x)
y9 = np.exp(x)

fig1 , ax1 = plt.subplots(nrows = 3,ncols = 3,figsize = (20,20))
ax1[0,0].plot(x,y1,"tab:blue")
ax1[0,0].set_title("Square Function")
ax1[0,0].set_xlabel(r'$X-Axis$',fontsize = 15)
ax1[0,0].set_ylabel(r'$Y-axis',fontsize = 15)

ax1[0,1].plot(x,y2,"tab:orange")
ax1[0,1].set_title("Cubic Function")
ax1[0,1].set_xlabel(r'$X-Axis$',fontsize = 15)
ax1[0,1].set_ylabel(r'$Y-axis',fontsize = 15)

ax1[0,2].plot(x,y3,"tab:green")
ax1[0,2].set_title("Sine Function")
ax1[0,2].set_xlabel(r'$X-Axis$',fontsize = 15)
ax1[0,2].set_ylabel(r'$Y-axis',fontsize = 15)

ax1[1,0].plot(x,y4,"b-")
ax1[1,0].set_title("Cosine Function")
ax1[1,0].set_xlabel(r'$X-Axis$',fontsize = 15)
ax1[1,0].set_ylabel(r'$Y-axis',fontsize = 15)

ax1[1,1].plot(x,y5,"r-")
ax1[1,1].set_title("Square Function")
ax1[1,1].set_xlabel(r'$X-Axis$',fontsize = 15)
ax1[1,1].set_ylabel(r'$Y-axis',fontsize = 15)

ax1[1,2].plot(x,y6,"g-")
ax1[1,2].set_title("Square Function")
ax1[1,2].set_xlabel(r'$X-Axis$',fontsize = 15)
ax1[1,2].set_ylabel(r'$Y-axis',fontsize = 15)

ax1[2,0].plot(x,y7,"y-")
ax1[2,0].set_title("Square Function")
ax1[2,0].set_xlabel(r'$X-Axis$',fontsize = 15)
ax1[2,0].set_ylabel(r'$Y-axis',fontsize = 15)

ax1[2,1].plot(x,y8,"r-")
ax1[2,1].set_title("Square Function")
ax1[2,1].set_xlabel(r'$X-Axis$',fontsize = 15)
ax1[2,1].set_ylabel(r'$Y-axis',fontsize = 15)

ax1[2,2].plot(x,y9,"g-")
ax1[2,2].set_title("Square Function")
ax1[2,2].set_xlabel(r'$X-Axis$',fontsize = 15)
ax1[2,2].set_ylabel(r'$Y-axis',fontsize = 15)
```
