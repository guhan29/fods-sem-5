```python
df = pd.DataFrame({
    'user_id':[1,2,1,3,3,],
    'content_id':[1,1,2,2,2],
    'tag':['cool','nice','clever','clever','not-bad']
})
gg=df.groupby('content_id')['tag'].apply(lambda tags:','.join(tags))
gg
```

```python
df.groupby('content_id')['user_id'].nunique()
```

```python
df.groupby('content_id')['user_id'].nunique().to_frame()
```

```python
df.groupby('content_id')['user_id'].nunique().to_frame().reset_index()
```

```python
df.groupby('content_id')['user_id'].nunique().to_frame().reset_index(drop=True)
```

```python
df = pd.DataFrame({
    'value':[20.45,22.89,32.12,111.22,33.22,100.00,99.99],
    'product':['table','chair','chair','mobile phone','table','mobile phone','table']
})
ee=df.groupby('product')['value'].sum().to_frame().reset_index().sort_values(by='value')
ee
```

```python
import matplotlib.pyplot as p
df.groupby('product').plot()
```

```python
df.groupby('product').size().plot(kind='bar')
```

```python
df = pd.DataFrame({
    'value':[20.45,22.89,32.12,111.22,33.22,100.00,99.99],
    'product':['table','chair','chair','mobile phone','table','mobile phone','table'],
    'price':[100,200,300,400,500,600,700]
})
group=df.groupby('product').agg({'value':['min','max','mean']})
group
```

```python
gro=df.groupby('product').agg({'value':['min','max'],'price':['mean']})
gro
```

```python
df.groupby('product')['value'].apply(lambda x:x).to_frame().reset_index(name='custom')
```

```python
gro.columns=['_'.join(col) for col in gro.columns.values]
gro=gro.reset_index()
gro
```
