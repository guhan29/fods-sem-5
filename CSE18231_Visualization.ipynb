```
import pandas as pd 
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import  scipy as sc
%matplotlib inline
sns.set_style('white')
```

```
i=sns.load_dataset('iris')
i['species']="Iris-" + i['species']
i.rename(columns={'sepal_length':'SepalLengthCm','sepal_width':'SepalWidthCm','petal_length':'PetalLengthCm','petal_width':'PetalWidthCm','species':'Species'},inplace=True)
i.head()
```

# 1

```
sns.jointplot(i.SepalLengthCm,i.SepalWidthCm,stat_func=sc.stats.pearsonr);
```

# 2

```
plt.figure(figsize=(4.5,4.5))
sns.scatterplot(i.SepalLengthCm,i.SepalWidthCm,hue=i.Species)
sns.despine()
plt.legend(bbox_to_anchor=(1,.5));
```

# 3

```
sns.boxplot(i.Species,i.PetalLengthCm);
```

# 4

```
sns.stripplot(i.Species,i.SepalLengthCm);
```

# 5

```
sns.boxplot(i.Species,i.SepalLengthCm)
sns.stripplot(i.Species,i.SepalLengthCm);
```

# 6

```
sns.violinplot(i.Species,i.SepalLengthCm);
```

# 7

```
sns.pairplot(i,diag_kind='hist',hue='Species');
```

# 8

```
plt.figure(figsize=(10,7))
sns.heatmap(i.corr(),annot=True,cmap='summer')
plt.xticks(rotation=0)
plt.yticks(rotation=0);
```

# 9

```
i.hist(edgecolor='black', linewidth=1);
plt.gcf().set_size_inches(10,6)
```

# 10

```
sns.set(style="whitegrid")
sns.swarmplot(i.Species,i.PetalLengthCm);
```
