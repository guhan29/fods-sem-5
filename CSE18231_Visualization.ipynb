```
import pandas as pd 
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import  scipy as sc
%matplotlib inline 
```

```
i=sns.load_dataset('iris')
sns.set_style('white')
i.head()
```

# 1

```
sns.jointplot(i.sepal_length,i.sepal_width,stat_func=sc.stats.pearsonr)
```

# 2

```
plt.figure(figsize=(4.5,4.5))
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
sns.scatterplot(i.sepal_length,i.sepal_width,hue=i.species)
```

```
sns.boxplot(i.species,i.petal_length);
```
